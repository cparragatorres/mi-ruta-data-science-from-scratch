# Formato con espacios en blanco

```python
# El signo de la almohadilla marca el comienzo de un comentario. Python
# ignora los comentarios, pero son √∫tiles para cualquiera que lea el c√≥digo.
for i in [1, 2, 3, 4, 5]:
  print(i)                    # primera l√≠nea del bloque "for i"
  for j in [1, 2, 3, 4, 5]:
    print(j)                  # primera l√≠nea del bloque "for j"
    print(i + j)              # √∫ltima l√≠nea del bloque "for j"
  print(i)                    # √∫ltima l√≠nea del bloque "for i"

print("done looping")
```

```python
def double(x):
  '''
  Aqu√≠ se coloca una cadena de documentaci√≥n opcional que explica la
  funci√≥n. Por ejemplo, esta funci√≥n multiplica su entrada por 2.
  '''
  return x * 2

double(5)  # Llamada a la funci√≥n, devuelve 10
```

```python
def apply_to_one(f):
  '''Llama a la funci√≥n f con 1 como argumento'''
  return f(1)

my_double = double            # se refiere a la funci√≥n anteriormente definida

x = apply_to_one(my_double)   # es igual a 2
x
```

```python
'''
Tambi√©n es f√°cil crear funciones an√≥nimas cortas, denominadas lambdas:
'''
# lambda argumentos: expresi√≥n
# Funci√≥n lambda que suma 2 n√∫meros
suma = lambda x, y: x + y

print(suma(5, 3))  # Resultado: 8

'''
Se pueden asignar lambdas a variables, aunque la mayor√≠a de la gente dir√°
que en su lugar solo hay que utilizar def:
'''
another_double = lambda x: 2 * suma # no haga esto
def another_double(suma):
  """Do this instead"""
  return 2 * suma

another_double(6)
```

```python
'''
Asignando argumentos predeterminados,
que solamente tienen que especificarse cuando se desea
un valor distinto al predeterminado
'''
def my_print(message = "my default message"):
  print(message)
my_print("hello") # imprime ‚Äòhello‚Äô
my_print() # imprime ‚Äòmy default message‚Äô


'''
Algunas veces es √∫til especificar argumentos por el nombre:
'''
def full_name(first = "What‚Äôs-his-name de", last = "quien-sabe"):
  return first + " " + last

full_name("Christian", "Parraga") # "Christian Parraga"
full_name("Christian") # "Christian quien-sabe"
full_name(last="Parraga") # "What‚Äôs-his-name de Parraga"
```

```python
"Las tuplas son una forma comoda de devolver varios valores de una funci√≥n"
def suma_y_producto(x, y):
  return (x + y), (x * y)  # devuelve una tupla

sp = suma_y_producto(2, 3)      # sp es (5, 6)
s, p = suma_y_producto(5, 10)   # s es 15, p es 50
```

```python
"Las tuplas (y las listas) se pueden utilizar para asignaci√≥n m√∫ltiple:"
x, y = 1, 2     # ahora x es 1, y es 2
print(x, y)  # Imprime: 2 1
x, y = y, x     # Forma pit√≥nica de intercambiar variables; ahora x es 2, y es 1
print(x, y)  # Imprime: 2 1
```

# Diccionarios

```python
'''Un diccionario es una colecci√≥n desordenada de pares clave-valor, asocia valores
a claves y permite recuperar r√°pidamente el valor correspondiente a una
determinada clave'''
claves = {"Joel": 80, "Tim": 95}  # dict literal

"Se puede consultar el valor para una clave utilizando corchetes:"
joel_valor = claves["Joel"]      # es igual a 80
print(joel_valor)

"Pero se obtendr√° un KeyError si se pregunta por una clave que no est√° en el diccionario:"

try:
  kate_valor = claves["Kate"]
except KeyError:
  print("sin valor para Kate!")
```

```python
"Se puede comprobar la existencia de una clave utilizando in:"
joel_tiene_clave = "Joel" in claves # True
kate_tiene_clave = "Kate" in claves # False
joel_tiene_clave, kate_tiene_clave

'''Los diccionarios tienen un metodo get que devuelve None (o un valor predeterminado que
especificamos) si la clave no est√° en el diccionario:'''
joel_valor = claves.get("Joel")         # es igual a 80
kates_valor = claves.get("Kate")        # es igual a None
kates_valor_predeterminado = claves.get("Kate", 0) # es igual a 0
no_ones_grade = claves.get("Ninguno") # el valor predeterminado es None
joel_valor, kates_valor, kates_valor_predeterminado, no_ones_grade
```

```python
"Se pueden asignar pares clave/valor utilizando los mismos corchetes:"
claves["Tim"] = 99            # reemplaza el valor anterior
claves["Kate"] = 100          # a√±ade una tercera entrada
num_estudiantes = len(claves) # es igual a 3
print(claves)
print(len(claves))
```

## defaultdict

```python
"Un defaultdict es un tipo especial de diccionario que crea valores",
'''La diferencia a un diccionario es que no da error cuando intentas acceder a una clave
que no existe: en lugar de eso, crea autom√°ticamente un valor por defecto.'''
"Para eso se debe importar el modulo collections y usar defaultdict:"
from collections import defaultdict

# Creamos un defaultdict que usa list() como valor por defecto
mi_diccionario = defaultdict(list)

mi_diccionario["frutas"].append("manzana")
print(mi_diccionario["frutas"])  # üëâ ['manzana']
print(mi_diccionario["verduras"])  # üëâ [] (se crea autom√°ticamente una lista vac√≠a)
mi_diccionario
```

```python
'''Se podr√≠a utilizar tambi√©n el sistema ‚Äúmejor pedir perd√≥n que permiso‚Äù y
simplemente manejar la excepci√≥n al intentar consultar una clave inexistente:'''
mi_diccionario = {}

try:
    # Intentamos agregar una fruta a la lista que est√° dentro del diccionario
    mi_diccionario["frutas"].append("manzana")
except KeyError:
    # Si la clave "frutas" no existe, se crea una lista vac√≠a
    mi_diccionario["frutas"] = []
    mi_diccionario["frutas"].append("manzana")

print(mi_diccionario)
# üëâ {'frutas': ['manzana']}
```

```python
'''Los defaultdict tambi√©n pueden ser muy √∫tiles cuando queremos que las claves
creen listas, diccionarios, o incluso valores personalizados autom√°ticamente.'''
from collections import defaultdict

# 1Ô∏è‚É£ Usando list() ‚Üí crea una lista vac√≠a por defecto
dd_list = defaultdict(list)
dd_list[2].append(1)
print(dd_list)
# üëâ {2: [1]}

# 2Ô∏è‚É£ Usando dict() ‚Üí crea un diccionario vac√≠o por defecto
dd_dict = defaultdict(dict)
dd_dict["Joel"]["City"] = "Seattle"
print(dd_dict)
# üëâ {'Joel': {'City': 'Seattle'}}

# 3Ô∏è‚É£ Usando una funci√≥n propia (lambda) ‚Üí crea una lista con dos ceros [0, 0]
dd_pair = defaultdict(lambda: [0, 0])
dd_pair[2][1] = 1
print(dd_pair)
# üëâ {2: [0, 1]}

'''Ser√°n √∫tiles cuando estemos utilizando diccionarios para ‚Äúrecopilar‚Äù
resultados seg√∫n alguna clave y no queramos comprobar todo el tiempo si la
clave sigue existiendo.'''
```

# Contadores

```python
'''Imagina que tienes una lista de cosas y quieres saber cu√°ntas veces se repite cada una.
Counter hace eso autom√°ticamente.
- Le das una lista.
- Te devuelve cu√°ntas veces aparece cada elemento.'''

# 1. Importamos Counter
from collections import Counter
# 2. Creamos una lista
numeros = [1, 2, 1, 3, 2, 1]
# 3. Contamos autom√°ticamente
conteo = Counter(numeros)
print(conteo)
# Resultado: Counter({1: 3, 2: 2, 3: 1})

'''Eso significa:
- El n√∫mero 1 aparece 3 veces
- El n√∫mero 2 aparece 2 veces
- El n√∫mero 3 aparece 1 vez  '''
```

```python
# üìù Ejemplo con palabras
from collections import Counter

palabras = ["hola", "mundo", "hola", "python", "hola"]

conteo_palabras = Counter(palabras)
print(conteo_palabras)
# Resultado: Counter({'hola': 3, 'mundo': 1, 'python': 1})
# ‚≠ê Obtener las m√°s comunes
for palabra, cantidad in conteo_palabras.most_common(2):
    print(palabra, cantidad)
# Resultado:
# hola 3
# mundo 1
```
